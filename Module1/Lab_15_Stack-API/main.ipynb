```python
from stackapi import StackAPI
SITE = StackAPI('stackoverflow')
comments = SITE.fetch('comments')
```

### Question 1: Find the questions and answers of last month.

```python
questions = SITE.fetch('questions', fromdate=1669933989, todate=1672525989,
    min=20, tagged='python', sort='votes')

questions
```

```python
answers = SITE.fetch('answers', fromdate=1669933989, todate=1672525989,
    min=20, tagged='python', sort='votes')

answers
```

### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'.

```python
the_most_voted = SITE.fetch('questions', fromdate=1674628389, todate=1674686278,
    min=5, tagged='python', sort='votes')

the_most_voted
```

### Question 3: Find the answers with id 6784 and 6473.

```python
user_list = [6784, 6473]

answers_id = SITE.fetch('users/{ids}/answers', ids = user_list)

print(len(answers_id['items']))
```
