```python
import pandas as pd
```

```python
d = {
    'TransactionID': [1, 2, 3, 4, 5],
    'ClientID': [231, 765, 453, 231, 892],
    'Profession': ['Self-employed', 'students', 'Horece', 'self_emlpoyed', 'finance'],
    'Bank_dep': ['009', '005', '007', '009', '003'],
    'Risk': ['high', 'high', 'medium', 'high', 'low'],
    'Number of credits': [2, 3, 5, 2, 3],
    'Revenue': [30200, 12700, 89400, 30200, 740000]
}

df = pd.DataFrame(data=d)
```

```python
df
```

```python
df1 = pd.DataFrame(columns={
                    'Method', 
                    'Library', 
                    'Code', 
                    'Result', 
                    'Encode'

})

df1.set_index(['Method'])
```

## 1. Dummy Encoding

```python
df_dum = pd.get_dummies(df)
df_dum
```

```python
#filling a table 
df1.loc[1]= ['Dummy','Dummy Encoding', 'Pandas', 'Numeric', 'get_dummies']
df1
```

## 2. Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()

#for profession because without order

df['LabelEncoding'] = labelencoder.fit_transform (df['Profession'])

df
```

```python
df1.loc[2]= ['Label','Label Encoding', 'sklearn', 'Numeric', 'fit_transform']
df1
```

## 3. Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder

#risk should be ranked ==> use ordinal encoder

#enc = OrdinalEncoder()
#df["Risk"] = enc.fit_transform(df["Risk"])
x = pd.DataFrame(df['Risk'])
ratings_order = ['low', 'medium', 'high']
oe = OrdinalEncoder(categories = [ratings_order], dtype=int)
risk_ord = oe.fit_transform(x)

df['risk_enc']=risk_ord
df
```

```python
df1.loc[3]= ['Ordinal','Ordinal Encoding', 'sklearn', 'Numeric', 'fit_transform']
df1
```
